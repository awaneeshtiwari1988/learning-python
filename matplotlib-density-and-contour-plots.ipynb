<h1>Density and contour plots</h1>

```python
# There are three contour plots that can be helpful for the task of displaying contour plots:

# plt.contour for contour plots

# plt.contourf for filled contour plots

# plt.imshow for showing images

%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-white')
import numpy as np
```

<h3>Visualizing a three dimensional function</h3>

```python
def f(x,y):
    return np.sin(x) ** 10 + np.cos(10 + y*x)*np.cos(x)
```

```python
# A contour plot can be created with the plt.contour function. 

# It takes three arguments: a grid of x values, a grid of y values, and a grid of z values. 

# The x and y values represent positions on the plot and the z values will be represented by the contour levels. 

x = np.linspace(0,5,50)
y = np.linspace(0,5,40)

X,Y = np.meshgrid(x,y)
Z = f(X,Y)
```

```python
# Standard line only contour plot

plt.contour(X,Y,Z, colors='black');
```

```python
# By default when a single color is is used , negative values are represented by dashed lines and positive values
# by solid lines. 

# The lines can be color coded by specifiying color map with cmap argument. 

plt.contour(X,Y,Z,20,cmap='RdGy');
```

```python
# Multiple range of colormaps are available which can be browsed using the plt.cm module. plt.cm.<TAB>
```

```python
# Filled color plot can be created using plt.contourf()

plt.contourf(X,Y,Z,20, cmap='RdGy')
plt.colorbar();
```

```python
plt.imshow(Z, extent=[0,5,0,5], origin='lower',cmap='RdGy')
plt.colorbar();
```

```python
contours = plt.contour(X,Y,Z,3,colors='black')
plt.clabel(contours,inline=True,fontsize=8)

plt.imshow(Z, extent=[0,5,0,5], origin='lower',cmap='RdGy',alpha=0.5)
plt.colorbar();
```
